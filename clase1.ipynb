```python
# get the security info from file
import json
keysAPI = json.load(open('claves.txt','r'))
```

```python
# credentials from https://apps.twitter.com/
consumerKey = keysAPI['consumer_key']
consumerSecret = keysAPI['consumer_secret']
accessToken = keysAPI['access_token']
accessTokenSecret = keysAPI['access_token_secret']
```

```python
# using security info:
import tweepy

auth = tweepy.OAuthHandler(consumerKey, consumerSecret)
auth.set_access_token(accessToken, accessTokenSecret)
api=tweepy.API(auth, wait_on_rate_limit=True,wait_on_rate_limit_notify=True,parser=tweepy.parsers.JSONParser())
```

```python
users.Cuenta[40]
```

```python
username=users.Cuenta[0]
userData = api.get_user(username)
```

```python
filas=[]
errores=[]
for user in users.Cuenta:
    try:
        allData=api.get_user(user)
        nombre=allData["name"]
        idt=allData["id"]
        seguidores=allData["followers_count"]
        amigos=allData["friends_count"]
        inicio=allData["created_at"]
        donde=allData["location"]
        filas.append([nombre, idt, seguidores, amigos, inicio, donde])
    except:
        errores.append(user)
    
```

```python
congre1=pd.DataFrame(filas,columns=["nombre", "idt", "seguidores", "amigos", "inicio", "donde"])
congre1.to_csv("congre1.csv",encoding="utf-8",index=False)
```
