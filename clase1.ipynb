#### csv

```python
import pandas

link="https://github.com/InvestigacionEscuelaGobPUCP/training/raw/master/idhPeru2012.csv"
data=pandas.read_csv(link, encoding='latin1',dtype={'UbiReg':str,'UbiPro':str,'UbiDis':str})
```

```python
data
```

```python
data.dtypes
```

```python
[len(x) for x in data.UbiReg]
```

```python
newvals1=['0'+ x if len(x)==5 else x for x in data.UbiReg ]
newvals2=['0'+ x if len(x)==5 else x for x in data.UbiPro ]
newvals3=['0'+ x if len(x)==5 else x for x in data.UbiDis ]
```

```python
data.UbiReg=newvals1
data.UbiPro=newvals2
data.UbiDis=newvals3
```

```python
import bs4
import pandas as pd

link2="https://en.wikipedia.org/wiki/Democracy_Index"
wiki=pd.read_html(link2,header=0,flavor='bs4',attrs = {'class': 'wikitable sortable'})[0]
```

```python
wiki
```

```python
wiki.dtypes
```

```python
wiki[wiki.Rank=="Rank"]
```

```python
wiki=wiki[wiki.Rank!="Rank"] # elimino esas filas
wiki
```

```python
wiki.Rank=wiki.Rank.str.replace("=","") #cambio =
wiki.Rank
```

```python
wiki.Score=[float(x) for x in wiki.Score]
```

```python
wiki.iloc[:,2:8]=wiki.iloc[:,2:8].applymap(float)
```
